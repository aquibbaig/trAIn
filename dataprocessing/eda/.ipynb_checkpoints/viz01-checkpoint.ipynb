```python
def dist_plot(data, col='all'):
    '''
    Draws Distribution plots for the column of the Dataframe
    '''
    all_numeric_columns = []
    for col in data.columns:
        if pd.api.types.is_numeric_dtype(data[col].dtype):
            all_numeric_columns.append(col)
            
    num_cols = len(all_numeric_columns)
    
    # initializing subplot instance
    fig, ax = plt.subplots(num_cols, figsize=(10, (num_cols)*5))
    
    # enumerating through the numeric cols
    for idx, col in enumerate(all_numeric_columns):  
        sns.distplot(data[col].dropna(), ax=ax[idx])
    
    return fig, ax
```

### Null Values - Heat Map

```python
def plot_null(data):
    
    if not isinstance(data, pd.DataFrame):
        raise Exception('data must be a pandas DataFrame')
    
    fig, ax = plt.subplots(figsize=(10,6))
    sns.heatmap(data.isnull(), ax=ax)
    
    return ax
```

```python
plot_null(df_titanic)
```
